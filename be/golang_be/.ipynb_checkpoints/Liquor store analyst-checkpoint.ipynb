```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_municipalities_in_British_Columbia')
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
name = [i.get_text().lower() for i in soup.find_all('td', scope = 'row')]
len(name)
```

```python
text = [ i.get_text() for i in soup.find_all('a') if len(i.get_text())>0]
txt = text[text.index('2016 Census of Population')+2:]
txt = txt[:txt.index('Zeballos')+2]
```

```python
text1= filter(lambda x: x[0]!= '[', txt)
regional_district = list(text1)[1::2]
len(regional_district)
```

```python
dt = pd.DataFrame({'name': name, 'regional_district': regional_district})
dt.head()
```

```python
lq.head()
```

```python
data = pd.merge(lq,dt , how = 'left', left_on ='city', right_on ='name')
data.head()
```

```python
def find_tags(var,begin,end):
    driver = webdriver.Chrome('/Users/NT/Downloads/chromedriver-3')
    price =[]
    name=[]
    page =[]
    country =[]
    sku_link = []
    SCROLL_PAUSE_TIME = 5
    for pagenum in tqdm(range(begin,end)):
        driver.get('https://www.bcliquorstores.com/product-catalogue?category=%s&sort=name.raw:asc&page=%d' %(var,pagenum))
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source,'html.parser')
        name_i = [i.get_text() for i in soup.find_all('h3', class_ = "product-name")]
        country_i = [i.get_text() for i in soup.find_all('span', class_ = "country-code")]
        sku_link_i = [i.get('href') for i in soup.find_all('a')][118:214:4]
        price_i= [i.get_text() for i in soup.find_all('div', class_ = "product-price")]
        #append to the final list
        name.append(name_i)
        country.append(country_i)
        sku_link.append(sku_link_i)
        price.append(price_i)
    print([len(i) for i in [name, country, sku_link, price]])
    return name
```
